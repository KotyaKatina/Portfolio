В части столбцов неверный тип данных: first_day_exposition - нужно изменить на datetime для удобства работы с датой. 
В столбцах floors_total и balcony должны быть целочисленные типы, так как они обозначают количество.

### Предобработка данных

Проверяем наличие пропущенных значений в столбцах таблицы:

```python
data.isna().sum()
```

В 14 столбцах присутствуют пропущенные значения.

В количестве балконов пропуски скорее всего встречаются, так у данных квартир нет балконов. Заполним такие пропуски нулями.

При заполнении информации о помещении те, чья квартира не является апартаментами, могли пропускать данный раздел, что привело к появлению Null. Целесообразно заполнить пропуски значением False.

Пропуски в высоте потолков заполняем медианным значением.

В остальных пропущенных значениях оставляем ячейки пустыми, заполнить их не представляется возможным, квартиры очень разные и заполнить средними или медианными значениями даже деля их на сегменты будет некорректно. Удалить такие строки не можем, так как их много и удаление может сильно исказить результаты анализа. Скорее всего заполняющие объявления люди не посчитали обязательным для продажи квартиры заполнение этих полей или не знали точных параметров.

Удаляем строки с пропущенными значениями в floors_total и locality_name, их отсутствие не повлияет значительно на результат, так как таких строк всего 86 и 49 соответственно.

```python
data['balcony'] = data['balcony'].fillna(0)

data['is_apartment'] = data['is_apartment'].fillna(False)

ceiling_height_median = data['ceiling_height'].median()
for ceiling_height_null in data['ceiling_height'].unique():
    data.loc[(data['ceiling_height'].isna()), 'ceiling_height'] = ceiling_height_median
data.info()

data = data.dropna(subset=['floors_total'])
data = data.dropna(subset=['locality_name'])

data.isna().sum()
```

Проверяем типы данных в датафрейме 

```python
data.info()
```

В столбце first_day_exposition изменим тип на datetime для удобства работы с датой.

В столбцах floors_total и balcony изменим тип на целочисленный, так как количество этажей и балконов не может быть выражено в дробном числе.

```python
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
data['floors_total']=data['floors_total'].astype('int')
data['balcony']=data['balcony'].astype('int')
data.info()
```

Проверяем наличие дубликаов в датафрейме

```python
data.duplicated().sum()
```

Изучаем столбец с названиями и удаляем неявные дубликаты

```python
data['locality_name'].sort_values().unique()
```

Избавляемся от длинных наименований и буквы "ё", для уменьшения повторов в названиях населенных пунктов:

```python
data['locality_name'] = data['locality_name'].str.replace('ё', 'е')
data['locality_name'] = data['locality_name'].str.replace('поселок городского типа', 'пгт')
data['locality_name'] = data['locality_name'].str.replace('деревня', 'д.')
data['locality_name'] = data['locality_name'].str.replace('поселок', 'п.')
data['locality_name'].sort_values().unique()
```

Заменяем видимые предположительные дубликаты в названиях:

```python
data['locality_name'] = data['locality_name'].str.replace('п. при железнодорожной станции Вещево', \
                                                          'п. станции Вещево', regex=True)
data['locality_name'] = data['locality_name'].str.replace('п. при железнодорожной станции Приветнинское', \
                                                          'п. станции Приветнинское', regex=True)
data['locality_name'] = data['locality_name'].str.replace('коттеджный п. Лесное', 'п. Лесное', regex=True)
```

Находим аномалии в столбце ceiling_height, если значения больше 10, разделим их на 10, так как скорее всего это опечатка.

```python
print(data['ceiling_height'].unique())
data.loc[data['ceiling_height'] > 10, 'ceiling_height'] = data['ceiling_height']/10
```

В размерах жилого помещения есть очень большие значения, но между средним и медианой разница не значительно отличается.

```python
print (data['last_price'].sort_values().unique())
print ('среднее:', data['last_price'].mean().round())
print ('медиана:', data['last_price'].median())
data.sort_values('last_price', ascending = False).head(20)
```

Удаляем строки с общей площадью больше 200 м, стоимостью более 50 млн и количеством комнат больше 10.

```python
data = data[(data.total_area <= 200) & (data.last_price <= 50000000) & (data.rooms < 10)]
data.shape[0]/data2.shape[0]
```

В результате предобработки отброшено около 2% данных, это значит, что результат не исказится значительно. 

### Посчитаем и добавим в таблицу:
- цену одного квадратного метра price_for_m,
- день недели публикации объявления published_day,
- месяц публикации объявления published_month,
- год публикации объявления published_year,
- тип этажа квартиры (значения — «первый», «последний», «другой»);
- расстояние до центра города в километрах cityCenters_nearest_km.

```python
data['price_for_m'] = data['last_price']/data['total_area']
data['published_day'] = data['first_day_exposition'].dt.weekday
data['published_month'] = pd.DatetimeIndex(data['first_day_exposition']).month
data['published_year'] = pd.DatetimeIndex(data['first_day_exposition']).year

data['floors_type']=''
data.loc[data['floor'] == 1, 'floors_type'] = 'первый'
data.loc[(data['floors_total'] == data['floor']) & (data['floors_total'] != 1), 'floors_type'] = 'последний'
data.loc[data['floors_type'] =='', 'floors_type'] = 'другой'

data['cityCenters_nearest_km'] = round(data['cityCenters_nearest']/1000)

data
```
